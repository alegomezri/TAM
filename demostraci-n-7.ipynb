
Sea $ X_n \sim \mathcal{N}(\mu, \sigma^2) $ y el estimador de mxima verosimilitud para la media, $ \mu_{ML}$, est dado por:

$
\mu_{ML} = \frac{1}{N} \sum_{n=1}^{N} X_n
$

Calculamos la esperanza matemtica de $ \mu_{ML}$:

$
\mathbb{E}[\mu_{ML}] = \mathbb{E}\left[\frac{1}{N} \sum_{n=1}^{N} X_n\right] = \frac{1}{N} \sum_{n=1}^{N} \mathbb{E}[X_n]
$

Dado que $ X_n \sim \mathcal{N}(\mu, \sigma^2) $, sabemos que $\mathbb{E}[X_n] = \mu $ para todo $ n $. Por lo tanto:

$
\mathbb{E}[\mu_{ML}] = \frac{1}{N} \sum_{n=1}^{N} \mu = \frac{1}{N} \cdot N \cdot \mu = \mu
$

Esto demuestra que $ \mu_{ML} $ es un estimador insesgado de $\mu$


El estimador de mxima verosimilitud para la varianza, $ \sigma^2_{ML} $, est dado por:

$
\sigma^2_{ML} = \frac{1}{N} \sum_{n=1}^{N} (X_n - \mu_{ML})^2
$

Calculamos la esperanza de $\sigma^2_{ML} $. 

Primero, descomponemos $ X_n $ como:

$
X_n = \mu + Z_n
$

Donde $ Z_n \sim \mathcal{N}(0, \sigma^2) $

Entonces, $ \sigma^2_{ML}$ se puede escribir como:

$
\sigma^2_{ML} = \frac{1}{N} \sum_{n=1}^{N} \left( Z_n + (\mu - \mu_{ML}) \right)^2
$

Expandiendo el trmino:

$
\sigma^2_{ML} = \frac{1}{N} \sum_{n=1}^{N} \left[ Z_n^2 + 2Z_n(\mu - \mu_{ML}) + (\mu - \mu_{ML})^2 \right]
$

La esperanza de $ \sigma^2_{ML}$ ser:

$
\mathbb{E}[\sigma^2_{ML}] = \frac{1}{N} \sum_{n=1}^{N} \left[ \mathbb{E}[Z_n^2] + 2\mathbb{E}[Z_n]\mathbb{E}[(\mu - \mu_{ML})] + \mathbb{E}[(\mu - \mu_{ML})^2] \right]
$

Sabemos que $ \mathbb{E}[Z_n] = 0 $ y \$ \mathbb{E}[Z_n^2] = \sigma^2 $, por lo tanto:

$
\mathbb{E}[\sigma^2_{ML}] = \frac{1}{N} \sum_{n=1}^{N} \left[ \sigma^2 + \mathbb{E}[(\mu - \mu_{ML})^2] \right]
$

Dado que:

$
\mathbb{E}[(\mu - \mu_{ML})^2] = \frac{\sigma^2}{N}
$

Entonces:

$
\mathbb{E}[\sigma^2_{ML}] = \sigma^2 \left(1 - \frac{1}{N}\right) = \left(\frac{N-1}{N}\right) \sigma^2
$

Esto demuestra que $ \sigma^2_{ML} $ es un estimador sesgado de $ \sigma^2 $, ya que su valor esperado no es igual a $ \sigma^2 $, sino que es menor.


Para corregir el sesgo del estimador $( \sigma^2_{ML}$, necesitamos multiplicarlo por el inverso del factor $ \frac{N-1}{N} $. As, el estimador corregido sera:

$
\sigma^2_{\text{corregido}} = \frac{N}{N-1} \cdot \sigma^2_{ML} = \frac{1}{N-1} \sum_{n=1}^{N} (X_n - \mu_{ML})^2
$

Este estimador ahora es insesgado, es decir, $ \mathbb{E}[\sigma^2_{\text{corregido}}] = \sigma^2 $.
