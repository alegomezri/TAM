
Dado que $( X_n = A + W_n ), donde ( W_n \sim \mathcal{N}(\mu, \sigma^2) )$, tenemos los siguientes estimadores:

1. **Estimador 1:** 

   $\hat{A}_1 = \frac{1}{N} \sum_{n=1}^{N} X_n$


2. **Estimador 2:**

   
  $\hat{A}_2 = X_1$


Primero, calculamos la esperanza matemática de ambos estimadores para verificar si son insesgados.

$\mathbb{E}[\hat{A}_1] = \mathbb{E}\left[\frac{1}{N} \sum_{n=1}^{N} X_n \right] = \frac{1}{N} \sum_{n=1}^{N} \mathbb{E}[X_n]
]$

Dado que $( X_n = A + W_n ) y ( \mathbb{E}[W_n] = \mu )$:


$\mathbb{E}[X_n] = A + \mu$
Entonces:

$\mathbb{E}[\hat{A}_1] = \frac{1}{N} \sum_{n=1}^{N} (A + \mu) = A + \mu$


Por lo tanto, $( \hat{A}_1 )$ es un estimador insesgado si $( \mu = 0 )$


$\mathbb{E}[\hat{A}_2] = \mathbb{E}[X_1] = A + \mu$


Por lo tanto, $( \hat{A}_2 )$ también es un estimador insesgado si $( \mu = 0 )$


Ahora, calculamos la varianza de ambos estimadores:


$\text{Var}(\hat{A}_1) = \text{Var}\left(\frac{1}{N} \sum_{n=1}^{N} X_n \right) = \frac{1}{N^2} \sum_{n=1}^{N} \text{Var}(X_n)$

Dado que $( \text{Var}(X_n) = \sigma^2)$:

$\text{Var}(\hat{A}_1) = \frac{1}{N^2} \cdot N \cdot \sigma^2 = \frac{\sigma^2}{N}$



$\text{Var}(\hat{A}_2) = \text{Var}(X_1) = \sigma^2$


Para determinar cuál estimador es mejor, comparamos las varianzas:

- $(\text{Var}(\hat{A}_1) = \frac{\sigma^2}{N})$
- $(\text{Var}(\hat{A}_2) = \sigma^2)$

Dado que $( \frac{\sigma^2}{N} \leq \sigma^2 )$ (con $( N \geq 1 )),$ la varianza de $( \hat{A}_1 )$ es menor que la varianza de $( \hat{A}_2 )$


El Estimador 1 $(\hat{A}_1)$ es mejor que el Estimador 2 $(\hat{A}_2)$ desde el punto de vista de la varianza, ya que tiene una varianza menor. Esto significa que $( \hat{A}_1 )$ es más eficiente y ofrece estimaciones más precisas de $( A )$.
