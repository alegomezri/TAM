```python
 #codificar variables moneda
def code_euro(y):
    yc = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
        if y[i][-1]=='M': yc[i] = float(y[i][1:-1])*10**6 #buscar M y reemplazar 10^6
        elif y[i][-1]=='K': yc[i] = float(y[i][1:-1])*10**3 # buscar K y reemplazar por 10^3
        else: yc[i] = float(y[i][1:])
    return yc

#codificar estadísticas
def code_stats(y):
    yc = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
        if y.iloc[i].find("+") > -1: # encontrar signo + en str y casteo a flotante
            yc[i] = float(y.iloc[i][:y.iloc[i].find("+")])+float(y.iloc[i][y.iloc[i].find("+")+1:])
        else: yc[i] = float(y.iloc[i])
    return yc
```

# Se crea clase propia (custom) para ejecutar el preproceso que sea compatible con scikitlearn

```python
#%% transformer custom
from sklearn.base import BaseEstimator, TransformerMixin #objetos básicos para herencia de clase

class mypre_fifa(BaseEstimator,TransformerMixin):
    #inicializacion de clase y varaibles
    def __init__(self, col_euro,col_stats, cat_usr): #constructor clase
        self.col_euro = col_euro #lista atributos tipo moneda
        self.col_stats = col_stats #lista atributos tipo estadistica
        self.cat_usr = cat_usr #lista de atributos categoricos

    def fit(self,X, *_):#función de ajuste -> *_ para indicar que puede recibir más entradas en el pipeline
        Xi = X.copy() #copiar dataset para no reemplazar original
        self.imputer_num = SimpleImputer(strategy="most_frequent") #crear imputador tipo moda
        self.a = Xi.columns[np.sum(Xi.isna())> 0] #encontrar columnas con datos faltantes
        self.imputer_num.fit(Xi[self.a]) # ajustar imputador
        Xi[self.a] = self.imputer_num.transform(Xi[self.a]) #evaluar datos con el imputador

        for i in self.col_euro: #codificar tipo moneda
            Xi[i] = code_euro(np.array(Xi[i]))

        for i in self.col_stats: #codificar datos estadisticos
            Xi[i] = code_stats(Xi[i])

        #height, wieght corregir formato
        Xi['Height'].replace(regex=["'"], value='.',inplace=True) #reemplaza unidad ' con .
        for i in Xi.index:
            Xi.loc[i,'Weight'] = float(Xi.loc[i,'Weight'][:-3])#elimnar unidades
            Xi.loc[i,'Height'] = float(Xi.loc[i,'Height'])#eliminar unidades

        Xi['Height'] = Xi['Height'].astype('float64')
        Xi['Weight'] = Xi['Weight'].astype('float64')#asegurar formato flotante

        Xi['Joined'] = Xi['Joined'].replace(regex="/",value="")#codificar fecha
        Xi['Joined'] = Xi['Joined'].astype('float64')

        cat = [] #codificar variables categoricas con ordinal encoder
        for i in self.cat_usr.keys():
            cat = cat + [[*self.cat_usr.get(i)]]
        self.col_cat_usr = OrdinalEncoder(categories=cat) # OneHotEncoder(categories=cat,sparse=False)
        Xi[[*self.cat_usr.keys()]] =self.col_cat_usr.fit_transform(Xi[[*self.cat_usr.keys()]])
        #si utilizar OneHotEncoder debe concatenar las columnas resultantes -> ver cuaderno guia lado a lado

        return self

    def transform(self, X, *_):#funcion transformador-> pensando en datos nuevos
        Xi = X.copy()
        Xi[self.a] = self.imputer_num.transform(Xi[self.a])

        for i in self.col_euro:
            Xi[i] = code_euro(np.array(Xi[i]))
        for i in self.col_stats:
            Xi[i] = code_stats(Xi[i])

                #height, wieght
        Xi['Height'].replace(regex=["'"], value='.',inplace=True)
        for i in Xi.index:
            Xi.loc[i,'Weight'] = float(Xi.loc[i,'Weight'][:-3])
            Xi.loc[i,'Height'] = float(Xi.loc[i,'Height'])

        Xi['Height'] = Xi['Height'].astype('float64')
        Xi['Weight'] = Xi['Weight'].astype('float64')

        Xi['Joined'] = Xi['Joined'].replace(regex="/",value="")
        Xi['Joined'] = Xi['Joined'].astype('float64')

        Xi[[*self.cat_usr.keys()]] =self.col_cat_usr.transform(Xi[[*self.cat_usr.keys()]])
        return Xi

    def fit_transform(self,X,*_):#ajustar y transformar en cascada
        self.fit(X)
        return self.transform(X)
```

```python
mypre = mypre_fifa(col_euro=col_euro,col_stats=col_stats,cat_usr = cat_usr)
Xtrain_pre = mypre.fit_transform(Xtrain)
```

**2.Modelos y Problemas de Optimización**

* Linear Regressor:
 
El modelo de regresión lineal es de la forma:

$y = X \beta + \epsilon$

donde $y$ son las variables de respuesta, $X$ es la matriz de variables explicativas (o características), $\beta$ son los coeficientes que queremos ajustar, y $\epsilon$ es el término de error.

**Problema de optimización**: Minimizar la suma de los errores cuadráticos medios.

$\min_{\beta} \left( \frac{1}{2} \| X \beta - y \|_2^2 \right)$

Es una optimización convexa, resuelta con métodos analíticos o algoritmos como la descomposición de Cholesky o la inversión de matrices.

* Lasso (Least Absolute Shrinkage and Selection Operator):

Similar a la regresión lineal, pero con un término de regularización $\ell_1$, lo que favorece soluciones más esparsas.

$
y = X \beta + \epsilon
$

**Problema de optimización:** Minimizar la suma de los errores cuadráticos medios con un término de penalización $\ell_1$.

$
\min_{\beta} \left( \frac{1}{2} \| X \beta - y \|_2^2 + \alpha \| \beta \|_1 \right)
$

El término de regularización $\alpha$ controla la magnitud de la penalización. Este problema se resuelve generalmente con algoritmos como \textit{Coordinate Descent} o \textit{Least Angle Regression}.

* ElasticNet

**Modelo**: Combina las penalizaciones $\ell_1$ (Lasso) y $\ell_2$ (Ridge).

$
y = X \beta + \epsilon
$

**Problema de optimización**: Minimizar la suma de los errores cuadráticos medios con una combinación de penalizaciones $\ell_1$ y $\ell_2$.

$
\min_{\beta} \left( \frac{1}{2} \| X \beta - y \|_2^2 + \alpha_1 \| \beta \|_1 + \alpha_2 \| \beta \|_2^2 \right)
$
Se usa para manejar la colinealidad y seleccionar variables. El problema se resuelve generalmente con \textit{Coordinate Descent}.

* Kernel Ridge

**Modelo**: Extensión de la regresión Ridge, utilizando un \textit{kernel} para transformar los datos a un espacio de mayor dimensión, permitiendo la captura de relaciones no lineales.

$
y = K \alpha + \epsilon
$

donde $K$ es la matriz kernel y $\alpha$ son los coeficientes.

**Problema de optimización**: Minimizar el error cuadrático medio con regularización $\ell_2$ en el espacio de características transformado por el kernel.

$
\min_{\alpha} \left( \| K \alpha - y \|_2^2 + \lambda \alpha^T K \alpha \right)
$
Este problema es resuelto generalmente mediante la inversión de matrices.

* SGDRegressor (Stochastic Gradient Descent Regressor)

**Modelo:** Utiliza el mismo modelo que la regresión lineal o modelos regulares (como Ridge o ElasticNet), pero en lugar de resolver el problema con métodos tradicionales, utiliza Stochastic Gradient Descent para optimizar los parámetros.

**Problema de optimización**: Minimiza la función objetivo (como en regresión lineal, Lasso, o ElasticNet) mediante pasos iterativos de gradiente descendente en minibatches de datos.

$
\min_{\beta} \left( \frac{1}{2} \| X \beta - y \|_2^2 \right) \text{ con actualizaciones iterativas}
$
Es particularmente útil cuando se tiene un conjunto de datos grande.

* Bayesian Ridge:

**Modelo**: Introduce un enfoque Bayesiano a la regresión lineal. Se asume que los coeficientes $\beta$ siguen una distribución normal con varianza desconocida, lo que da lugar a una distribución a posteriori sobre $\beta$ dada por los datos.

$
p(\beta | X, y) \propto p(y | X, \beta) p(\beta)
$

**Problema de optimización**: En lugar de minimizar una función de error directamente, el enfoque Bayesiano encuentra la distribución a posteriori de los coeficientes. La optimización se refiere a maximizar la probabilidad marginal o "evidencia".

$
\text{Maximizar } p(y | X)
$

Se resuelve usando técnicas Bayesianas, como la descomposición variacional.

* Gaussian Process Regressor

**Modelo**: Es un modelo no paramétrico que asume que las observaciones $y$ siguen una distribución conjunta Gaussiana. Los valores de $y$ se modelan como realizaciones de un proceso estocástico definido por una función de covarianza o kernel.

$
p(y | X) \sim \mathcal{N}(\mu(X), K(X, X'))
$

**Problema de optimización**: El ajuste del modelo implica maximizar la probabilidad de los datos observados, o equivalentemente minimizar la log-verosimilitud negativa.

$
\min_{\theta} \log p(y | X, \theta)
$

El parámetro $\theta$ corresponde a los parámetros del kernel, y este problema generalmente se resuelve mediante técnicas de optimización numérica como BFGS.


```python
#Hiperparametros:

params_lasso = {'regresion__alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

params_elastic = {'regresion__alpha': [0.001, 0.01, 0.1, 1, 10],
                  'regresion__l1_ratio': [0.1, 0.5, 0.9]}

params_krr = {'regresion__alpha': [0.1, 1, 10],
              'regresion__gamma': [0.001, 0.01, 0.1, 1]}

params_sgd = {'regresion__alpha': [0.0001, 0.001, 0.01, 0.1],
              'regresion__penalty': ['l2', 'l1', 'elasticnet']}

params_bayesian = {'regresion__alpha_1': [1e-6, 1e-5, 1e-4],
                   'regresion__alpha_2': [1e-6, 1e-5, 1e-4]}

params_gpr = {'regresion__alpha': [1e-10, 1e-5, 1e-2],
              'regresion__kernel': [RBF(length_scale=l) for l in [1, 10, 100]]}
```

```python
# Definir los diferentes regresores (solo los que quieres evaluar en el conjunto completo)
regressors = {
    'LinearRegression': (LinearRegression(), {}),
    'Lasso': (Lasso(), params_lasso),
    'ElasticNet': (ElasticNet(), params_elastic),
    'SGDRegressor': (SGDRegressor(), params_sgd),
    'BayesianRidge': (BayesianRidge(), params_bayesian),
}

# Crear el pipeline base (solo con el scaler)
steps = [('scaler', StandardScaler())]

# Evaluar cada regresor en el conjunto completo de datos
for name, (regressor, param_grid) in regressors.items():
    print(f"Evaluando: {name}")
    
    # Agregar el regresor al pipeline
    steps_with_regressor = steps + [('regresion', regressor)]
    pipeline = Pipeline(steps=steps_with_regressor)
    
    # Realizar el grid search
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_absolute_error')
    grid_search.fit(Xtrain_pre, ytrain_num)
    
    # Mejor modelo
    print(f"Mejor parámetro para {name}: {grid_search.best_params_}")
    print(f"Mejor MAE para {name}: {-grid_search.best_score_}")
    
    # Evaluar en el conjunto de test
    ytest_pred = grid_search.predict(Xtest_pre)
    mae_test = mean_absolute_error(ytest_num, ytest_pred)
    mse_test = mean_squared_error(ytest_num, ytest_pred)
    print(f"MAE en conjunto de test para {name}: {mae_test}")
    print(f"MSE en conjunto de test para {name}: {mse_test}\n")


# Parte especial para Kernel Ridge y Gaussian Process Regressor

# Reducir el conjunto de entrenamiento a 5000 puntos
Xtrain_small = Xtrain_pre[:5000]
ytrain_small = ytrain_num[:5000]

# Ajustar el grid search para Kernel Ridge con función rbf
print("Evaluando: KernelRidge (rbf)")
grid_search_krr = GridSearchCV(Pipeline(steps + [('regresion', KernelRidge(kernel='rbf'))]), params_krr, cv=5, scoring='neg_mean_absolute_error')
grid_search_krr.fit(Xtrain_small, ytrain_small)

# Mejor modelo Kernel Ridge
print(f"Mejor parámetro para Kernel Ridge: {grid_search_krr.best_params_}")
print(f"Mejor MAE para Kernel Ridge: {-grid_search_krr.best_score_}")

# Evaluar en el conjunto de test (con el conjunto completo)
ytest_pred_krr = grid_search_krr.predict(Xtest_pre)
mae_test_krr = mean_absolute_error(ytest_num, ytest_pred_krr)
mse_test_krr = mean_squared_error(ytest_num, ytest_pred_krr)
print(f"MAE en conjunto de test para Kernel Ridge: {mae_test_krr}")
print(f"MSE en conjunto de test para Kernel Ridge: {mse_test_krr}\n")


# Ajustar el grid search para Gaussian Process Regressor
print("Evaluando: GaussianProcessRegressor")
grid_search_gpr = GridSearchCV(Pipeline(steps + [('regresion', GaussianProcessRegressor())]), params_gpr, cv=5, scoring='neg_mean_absolute_error')
grid_search_gpr.fit(Xtrain_small, ytrain_small)

# Mejor modelo Gaussian Process Regressor
print(f"Mejor parámetro para Gaussian Process Regressor: {grid_search_gpr.best_params_}")
print(f"Mejor MAE para Gaussian Process Regressor: {-grid_search_gpr.best_score_}")

# Evaluar en el conjunto de test (con el conjunto completo)
ytest_pred_gpr = grid_search_gpr.predict(Xtest_pre)
mae_test_gpr = mean_absolute_error(ytest_num, ytest_pred_gpr)
mse_test_gpr = mean_squared_error(ytest_num, ytest_pred_gpr)
print(f"MAE en conjunto de test para Gaussian Process Regressor: {mae_test_gpr}")
print(f"MSE en conjunto de test para Gaussian Process Regressor: {mse_test_gpr}\n")
```
